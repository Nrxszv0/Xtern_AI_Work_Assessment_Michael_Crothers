```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv') 
df.head()
```

```python
year_order_counts = df.groupby(['Year', 'Order']).size().unstack(fill_value=0)

# Plot the grouped bar chart
# plt.figure(figsize=(20, 6))
year_order_counts.plot(kind='bar', stacked=False, colormap='gist_rainbow', figsize=(15,6))
plt.title('Number of Orders for Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
# plt.legend(title='Order', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.legend(title='Order', bbox_to_anchor=(1.05, 1), loc='upper left')

major_order_counts = df.groupby(['Major', 'Order']).size().unstack(fill_value=0)

# Plot the grouped bar chart
# plt.figure(figsize=(20, 6))
major_order_counts.plot(kind='bar', stacked=False, colormap='gist_rainbow', figsize=(20,6))
plt.title('Number of Orders for Each Major')
plt.xlabel('Major')
plt.ylabel('Number of Orders')
plt.legend(title='Order', bbox_to_anchor=(1.05, 1), loc='upper left')

university_order_counts = df.groupby(['University', 'Order']).size().unstack(fill_value=0)

# Plot the grouped bar chart
# plt.figure(figsize=(20, 6))
university_order_counts.plot(kind='bar', stacked=False, colormap='gist_rainbow', figsize=(20,6))
plt.title('Number of Orders for Each University')
plt.xlabel('University')
plt.ylabel('Number of Orders')
plt.legend(title='Order', bbox_to_anchor=(1.05, 1), loc='upper left')

time_order_counts = df.groupby(['Time', 'Order']).size().unstack(fill_value=0)

# Plot the grouped bar chart
# plt.figure(figsize=(20, 6))
time_order_counts.plot(kind='bar', stacked=False, colormap='gist_rainbow', figsize=(20,6))
plt.title('Number of Orders for Each Time')
plt.xlabel('Time')
plt.ylabel('Number of Orders')
plt.legend(title='Order', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
year_counts = df['Year'].value_counts()
# year_counts = df.sort_values('Year')
# year_counts = year_counts['Year'].value_counts()

# Plot the bar graph
plt.figure(figsize=(6, 6))
year_counts.sort_values().plot(kind='bar', color='salmon')
plt.title('Number of Orders For Each Year')
plt.ylabel('Number of Orders')
plt.xlabel('Year')

major_counts = df['Major'].value_counts()

# Plot the bar graph
plt.figure(figsize=(6, 6))
major_counts.sort_values().plot(kind='bar', color='salmon')
plt.title('Number of Orders For Each Major')
plt.ylabel('Number of Orders')
plt.xlabel('Major')

university_counts = df['University'].value_counts()

# Plot the bar graph
plt.figure(figsize=(6, 6))
university_counts.sort_values().plot(kind='bar', color='salmon')
plt.title('Number of Orders For Each University')
plt.ylabel('Number of Orders')
plt.xlabel('University')

time_counts = df['Time'].value_counts()

# Plot the bar graph
plt.figure(figsize=(6, 6))
time_counts.sort_values().plot(kind='bar', color='salmon')
plt.title('Number of Orders For Each Time')
plt.ylabel('Number of Orders')
plt.xlabel('Time')

order_counts = df['Order'].value_counts()

# Plot the bar graph
plt.figure(figsize=(10, 6))
order_counts.sort_values().plot(kind='bar', color='salmon')
plt.title('Number of Orders for Each Item')
plt.ylabel('Number of Orders')
plt.xlabel('Order')
plt.show()
```
